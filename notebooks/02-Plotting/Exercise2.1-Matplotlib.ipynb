# Plotting with Matplotlib

[Matplotlib](https://matplotlib.org/) is one of the most used plotting libraries in Python. It can be used to produce publication quality figures in a variety of formats, and is very flexible.

It can also be integrated with Jupyter, and it can work in different ways.

We will start by using the `inline` backend. Later you should change this to `widget`, and restart the kernel.

```python
%matplotlib inline
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 3*np.pi, 500)
fig, ax = plt.subplots()
plt.plot(x, np.sin(x**2))
plt.title('A simple chirp');
```
